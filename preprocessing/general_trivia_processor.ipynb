```python
from datasets import load_dataset

ds = load_dataset("mandarjoshi/trivia_qa", "rc")
print(ds)
```

```python
import pandas as pd

train_df = ds['train'].to_pandas()
# validation_df = ds['validation'].to_pandas()
# test_df = ds['test'].to_pandas()
```

```python
train_df['question_id'] = ['1'+str(i).zfill(9) for i in range(len(train_df))]
train_df['answer'] = train_df['answer'].apply(lambda x: x['value'])

# train_df_adjusted = train_df.drop(columns=['search_results', 'entity_pages', 'question_source'])

description_train_df = (
    train_df[['question_id', 'search_results']]
    .assign(search_results=lambda df: df['search_results'].apply(lambda x: x['search_context']))
    .explode('search_results', ignore_index=True)
    .dropna()
    .assign(description_id=lambda df: range(1, len(df) + 1))
)

url_train_df =  (
    train_df[['question_id', 'search_results']]
    .assign(search_results=lambda df: df['search_results'].apply(lambda x: x['url']))
    .explode('search_results', ignore_index=True)
    .dropna()
    .assign(url_id=lambda df: range(1, len(df) + 1))
)

answers_train_df = train_df[['question_id', 'answer']]
```

```python
# create questions csv

questions_general_df = train_df[['question', 'question_id', 'answer']]
questions_general_df.loc[:, 'category'] = ' '
questions_general_df.loc[:, 'jeopardy_or_general'] = 1

questions_jeopardy_df = pd.read_csv('updated_jeopardy_1.csv', dtype={'index': str})
questions_jeopardy_df['question_id'] = questions_jeopardy_df['index']
questions_jeopardy_df['question'] = questions_jeopardy_df[' Question']
questions_jeopardy_df['answer'] = questions_jeopardy_df[' Answer']
questions_jeopardy_df['category'] = questions_jeopardy_df[' Category']
questions_jeopardy_df = questions_jeopardy_df[['question', 'question_id', 'answer', 'category']]
questions_jeopardy_df.loc[:, 'jeopardy_or_general'] = 0
```

```python
# Concatenate the two DataFrames
questions_combined_df = pd.concat([questions_general_df, questions_jeopardy_df], axis=0, ignore_index=True).sort_values('question_id')
```
